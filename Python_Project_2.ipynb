```python
#loading dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv')
```

```python
df
```

```python
#data cleaning to remove unknown gender input
df = df[df.gender != 'Unknown/Invalid']
```

```python
df
```

```python
print(df.columns)
```

```python
print(df.info())
```

```python
#count and group genders by age
age_gender_counts = df.groupby(['age', 'gender']).size().unstack()
```

```python
print(age_gender_counts)
```

```python
#to create a bar chart for patients with DM by age group
plt.figure(figsize=(10, 5))
age_gender_counts.plot(kind='bar', stacked=False, figsize=(10, 5), colormap="coolwarm")
plt.title("Number of Patients with Diabetes in Each Age Group by Gender")
plt.xlabel("Age Group")
plt.ylabel("Number of Patients")
plt.legend(title="Gender")
plt.xticks(rotation=45)
plt.show()
```

```python
#remove unknown race value
df = df[df.race != '?']
```

```python
#count and group races by age
age_race_counts = df.groupby(['age', 'race']).size().unstack()
```

```python
print(age_race_counts)
```

```python
#to create a bar chart for patients with DM by race in each age group
plt.figure(figsize=(10, 5))
age_race_counts.plot(kind='bar', stacked=False, figsize=(10, 5), colormap="coolwarm")
plt.title("Number of Patients with Diabetes in Each Age Group by Race")
plt.xlabel("Age Group")
plt.ylabel("Number of Patients")
plt.legend(title="Race")
plt.xticks(rotation=45)
plt.show()
```

```python
conclusion = """The analysis shows that most diabetes patients fall into older age groups such as age 60 through 90.
Both male and female patients experience a similar increase in diabetes diagnosis once they reach the age of 50.
However, the number of female patients is slightly higher in older age groups compared to male patients.
When considering race, there is a noticeable variation in the diagnosis of diabetes across different racial groups.
While Caucasians represent a significant portion of diabetes patients across all age groups, African American and Hispanic patients show
a higher count of patients in the 50-70 age range. The diagnosis of diabetes appears to be more apparent in older patients,
regardless of race, but some racial groups tend to have a higher number of diagnoses in younger age groups compared to others."""
print(conclusion)
```
